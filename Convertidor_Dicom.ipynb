```python
import pydicom
import numpy as np
import cv2

# Cargar el archivo DICOM
dicom_file = "/content/1-1.dcm"  # Ruta del archivo DICOM
dicom_data = pydicom.dcmread(dicom_file)

# Convertir a una imagen en escala de grises (los datos están en dicom_data.pixel_array)
image_array = dicom_data.pixel_array.astype(np.uint8)

# Guardar como PNG
cv2.imwrite("imagen.png", image_array)

print("Conversión completa: imagen.png guardada.")
```
